# **Guassian Mixture Models** 

```python
import  numpy as np 
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs 
import matplotlib.pyplot as plt 
from sklearn import datasets 

## Load the Iris Dataset
iris  = datasets.load_iris()
X= iris.data[:,:2] # Taking only the first two features for visualization

## Creating the Guassian Mixture Model 

gmm = GaussianMixture(n_components=3)

## Fitting the model to the data 
gmm.fit(X)

## Fitting the cluster  labels for the data 
labels = gmm.predict(X)

## Visualizing the clusters through the diagram 
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis')
plt.xlabel('Sepal Lenghth')
plt.ylabel('Sepal Width ')
plt.title('Gaussian Mixture Model Clusterring on Iris Dataset')
plt.show()



```
