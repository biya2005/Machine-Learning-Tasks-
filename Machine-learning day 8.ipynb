### Logistics regression is checked when the problem of classification comes up. In regression we basically predict the numeric value. Logistics regression is one which is used to predict the classification data. We classify it after reading the data into binary classes(True and False) and the other is multi class classifiaction(not in binary, more than one )

## In the classification of logistics regression. The following metrics are used which are quite important.
### 1- Accuracy score 
### 2- Recall score 
### 3- Precision score
### 4- F1 score 
### 5- Confusion metrics 


### Logistic Regression: It predicts which means proability the occurrence of an event. If it is 0 means does not occur and if 1 it means it highly occurs.It is proably equals to 0.5.

### Sgmoid Function is used in the logistics regression which works as a converstion of numeric value into binary value either in 0 or 1. 


### Assumption of logistic regression  
### 1- The dependent variable should be binary in nature.
### 2- The independent variables should be continuous in nature.
### 3- The relationship between the independent and dependent variable should be non-linear.
### 4- There should not be outliers in the data. Check the outliers
